```python
#Q1. What is hypothesis testing in statistics
#Ans: Hypothesis testing is a statistical method that uses sample data to evaluate a claim (hypothesis) about a population parameter, determining if there's enough evidence to reject a null hypothesis.
```

```python
#Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis
#Ans: The null hypothesis (H₀) is a statement of no effect or no difference, while the alternative hypothesis (H₁) proposes that there is a statistically significant effect or difference.
```

```python
#Q3. What is the significance level in hypothesis testing, and why is it important
#Ans: The significance level (α) in hypothesis testing is the pre-determined probability of rejecting the null hypothesis when it is actually true (a Type I error); it's important because it sets the threshold for deciding whether the observed data provides enough evidence to reject the null hypothesis and thus controls the risk of making a false positive conclusion.
```

```python
#Q4. What does a P-value represent in hypothesis testing?
#Ans: The P-value represents the probability of observing a test statistic as extreme as, or more extreme than, the one calculated 1  from the sample data, assuming the null hypothesis is true
```

```python
#Q5. How do you interpret the P-value in hypothesis testing?
#Ans: A smaller P-value indicates stronger evidence against the null hypothesis, suggesting that the observed data is unlikely to 1  have occurred if the null hypothesis were true, thus potentially leading to its rejection if the P-value is below the pre-determined significance level (α).
```

```python
#Q6.  What are Type 1 and Type 2 errors in hypothesis testing?
#Ans: Type I error (false positive) occurs when we reject a true null hypothesis, while Type II error (false negative) occurs when we fail to reject a false null hypothesis
```

```python
#Q7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
#Ans: A one-tailed test examines if the population parameter is greater than or less than a specific value, while a two-tailed test examines if the population parameter is different from a specific value (it can be greater or less)
```

```python
#Q8. What is the Z-test, and when is it used in hypothesis testing?
#Ans: The Z-test is a statistical hypothesis test used to determine if there is a significant difference between a sample mean and a 1  population mean (when the population standard deviation is 2  known) or between the means of two independent samples (when both population standard deviations are known), and it is typically used when the sample size is large (n ≥ 30) due to the Central Limit Theorem
```

```python
#Q9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
#Ans:The Z-score is calculated as z= x-μ/σ representing the number of standard deviations the sample mean (x) is from the population mean (μ) under the null hypothesis
```

```python
#Q10. What is the T-distribution, and when should it be used instead of the normal distribution?
#Ans: The T-distribution is a probability distribution similar to the normal distribution but with heavier tails, used for small sample sizes or when the population standard deviation is unknown.
```

```python
#Q11. A What is the difference between a Z-test and a T-test?
#Ans: The key difference lies in the knowledge of the population standard deviation: a Z-test is used when the population standard deviation is known, while a T-test is used when the population standard 1  deviation is unknown and estimated from the sample
```

```python
#Q12. What is the T-test, and how is it used in hypothesis testing?
#Ans: The T-test is a statistical test used to determine if there is a significant difference between the means of two groups, 1  especially when the population standard deviation is unknown and estimated from the sample 
```

```python
#Q14. What is a confidence interval, and how is it used to interpret statistical results?
#Ans: A confidence interval is a range of values that is likely to contain the true value of a population parameter with a certain 1  degree of confidence.
```

```python
#Q15.  Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm

def z_test(sample_data, pop_mean, pop_std):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Calculate Z-statistic
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    
    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))
    
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z:.2f}")
    print(f"P-value: {p_value:.4f}")
    
    # Decision
    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis (significant result).")
    else:
        print("Fail to reject the null hypothesis (not significant).")

np.random.seed(0)
sample = np.random.normal(loc=102, scale=15, size=30)  # Sample with unknown mean
z_test(sample_data=sample, pop_mean=100, pop_std=15)   # Test against population mean = 100
```

```python
#Q16. Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    margin = std_err * stats.t.ppf((1 + confidence) / 2, df=n-1)
    lower = mean - margin
    upper = mean + margin

    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    return lower, upper

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Sample from normal distribution
confidence_interval(sample_data)
```

```python
#Q1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
import numpy as np
from scipy import stats

def z_test(sample_data, population_mean, population_std, sample_size):
    sample_mean = np.mean(sample_data)
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    
    # Calculate the two-tailed p-value
    p_value = 2 * (1 - stats.norm.cdf(np.abs(z_statistic)))
    
    return z_statistic, p_value

# Step 1: Generate a sample dataset (sample size = 50)
np.random.seed(42)
population_mean = 100  # Known population mean
population_std = 15    # Known population standard deviation
sample_size = 50
sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Step 2: Perform the Z-test
z_statistic, p_value = z_test(sample_data, population_mean, population_std, sample_size)

# Step 3: Output the results
print(f"Sample Mean: {np.mean(sample_data):.2f}")
print(f"Z-statistic: {z_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Interpret the results (using alpha = 0.05 as a typical threshold)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample mean and population mean.")
```

```python
#Q5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def hypothesis_testing_with_errors(sample_data, population_mean, population_std, sample_size, alpha=0.05):
    # Step 1: Perform Z-test for the sample mean
    sample_mean = np.mean(sample_data)
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    
    # Calculate two-tailed p-value
    p_value = 2 * (1 - stats.norm.cdf(np.abs(z_statistic)))
    
    # Step 2: Calculate critical value for Type I error (alpha)
    critical_value = stats.norm.ppf(1 - alpha / 2)
    
    # Step 3: Calculate Type I error (alpha)
    type_1_error = alpha
    
    # Step 4: Calculate Type II error (beta)
    # We assume that the true population mean is shifted (alternative hypothesis)
    true_population_mean = population_mean + 5  # Assume true population mean is shifted by 5
    z_beta = (critical_value - (true_population_mean - population_mean) / (population_std / np.sqrt(sample_size)))
    type_2_error = stats.norm.cdf(z_beta) - stats.norm.cdf(-z_beta)
    
    # Step 5: Visualize the decision regions
    # Generate the z-distribution curve
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x, loc=0, scale=1)

    plt.plot(x, y, label="Standard Normal Distribution", color='blue')
    
    # Shade the critical regions for Type I error (alpha)
    plt.fill_between(x, 0, y, where=(x <= -critical_value) | (x >= critical_value), color='red', alpha=0.3, label="Rejection Region (Type I Error)")
    
    # Shade the region of Type II error (beta)
    plt.fill_between(x, 0, y, where=(x > -critical_value) & (x < critical_value), color='yellow', alpha=0.3, label="Fail to Reject (Type II Error)")
    
    # Mark the Z-statistic and critical value on the plot
    plt.axvline(z_statistic, color='green', linestyle='dashed', label=f"Z-statistic: {z_statistic:.2f}")
    plt.axvline(-z_statistic, color='green', linestyle='dashed')
    plt.axvline(critical_value, color='orange', linestyle='dotted', label=f"Critical Value: {critical_value:.2f}")
    plt.axvline(-critical_value, color='orange', linestyle='dotted')
    
    # Labels and title
    plt.title('Hypothesis Testing: Type I and Type II Errors')
    plt.xlabel('Z-Score')
    plt.ylabel('Probability Density')
    plt.legend(loc='best')
    
    # Show the plot
    plt.show()

    # Step 6: Output the results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z_statistic:.2f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Type I Error Probability (alpha): {type_1_error:.4f}")
    print(f"Type II Error Probability (beta): {type_2_error:.4f}")
    
    # Interpretation
    if p_value < alpha:
        print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis: No significant difference between the sample mean and population mean.")

# Example usage:
np.random.seed(42)
population_mean = 100  # Known population mean
population_std = 15    # Known population standard deviation
sample_size = 50       # Sample size
sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Perform hypothesis testing and visualize Type I and Type II errors
hypothesis_testing_with_errors(sample_data, population_mean, population_std, sample_size)
```

```python
#Q9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance
import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(sample, confidence_level=0.95):
    # Sample statistics
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1) 
    sample_size = len(sample)
    
    # Degrees of freedom
    df = sample_size - 1
    
    # t-value for the given confidence level
    t_value = stats.t.ppf((1 + confidence_level) / 2, df)
    
    # Standard error of the mean
    standard_error = sample_std / np.sqrt(sample_size)
    
    # Margin of error
    margin_of_error = t_value * standard_error
    
    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Sample Standard Deviation: {sample_std:.4f}")
    print(f"Standard Error: {standard_error:.4f}")
    print(f"t-value: {t_value:.4f}")
    print(f"Margin of Error: {margin_of_error:.4f}")
    
    return lower_bound, upper_bound

# Example usage:
np.random.seed(42)
# Simulate a sample of 50 data points with mean 100 and standard deviation 15
sample_data = np.random.normal(loc=100, scale=15, size=50)

# Calculate the 95% confidence interval for the sample mean
lower, upper = calculate_confidence_interval(sample_data, confidence_level=0.95)
print(f"Confidence Interval: ({lower:.4f}, {upper:.4f})")
```

```python
#Q10. Write a Python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
import scipy.stats as stats

def calculate_margin_of_error(sample, confidence_level=0.95):
    # Sample statistics
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (using ddof=1 for sample)
    sample_size = len(sample)
    
    # Degrees of freedom
    df = sample_size - 1
    
    # t-value for the given confidence level
    t_value = stats.t.ppf((1 + confidence_level) / 2, df)
    
    # Standard error of the mean
    standard_error = sample_std / np.sqrt(sample_size)
    
    # Margin of error
    margin_of_error = t_value * standard_error
    
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Sample Standard Deviation: {sample_std:.4f}")
    print(f"Standard Error: {standard_error:.4f}")
    print(f"t-value: {t_value:.4f}")
    print(f"Margin of Error: {margin_of_error:.4f}")
    
    return margin_of_error

# Example usage:
np.random.seed(42)
# Simulate a sample of 50 data points with mean 100 and standard deviation 15
sample_data = np.random.normal(loc=100, scale=15, size=50)

# Calculate the margin of error for the sample mean with a 95% confidence level
margin_of_error = calculate_margin_of_error(sample_data, confidence_level=0.95)
print(f"Calculated Margin of Error: {margin_of_error:.4f}")
```

```python
#Q11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayesian_inference(prior, sensitivity, false_positive_rate):
    prior_no_disease = 1 - prior
    evidence = (sensitivity * prior) + (false_positive_rate * prior_no_disease)

    posterior = (sensitivity * prior) / evidence

    print(f"Prior (P(Disease)): {prior}")
    print(f"Sensitivity (P(Positive | Disease)): {sensitivity}")
    print(f"False Positive Rate (P(Positive | No Disease)): {false_positive_rate}")
    print(f"Posterior (P(Disease | Positive Test)): {posterior:.4f}")
    return posterior

# Example usage:
prior = 0.01  
sensitivity = 0.99  
false_positive_rate = 0.05

bayesian_inference(prior, sensitivity, false_positive_rate)
```

```python
#Q12. Perform a Chi-square test for independence between two categorical variables in Python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Create a sample dataset
data = {'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Preference': ['Tea', 'Coffee', 'Tea', 'Tea', 'Coffee', 'Coffee', 'Tea', 'Coffee']}

df = pd.DataFrame(data)

# Step 2: Create a contingency table
contingency_table = pd.crosstab(df['Gender'], df['Preference'])
print("Contingency Table:")
print(contingency_table)

# Step 3: Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Step 4: Display results
print("\nChi-Square Test Results:")
print(f"Chi2 Statistic: {chi2:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p:.4f}")

# Step 5: Interpret results
alpha = 0.05
if p < alpha:
    print("Result: Reject the null hypothesis — variables are dependent (associated).")
else:
    print("Result: Fail to reject the null hypothesis — variables are independent.")
```

```python
#Q16.  Implement an F-test using Python to compare the variances of two random samples
import numpy as np
from scipy.stats import f

# Step 1: Generate two random samples
np.random.seed(0)
sample1 = np.random.normal(loc=50, scale=10, size=30)  # std dev = 10
sample2 = np.random.normal(loc=50, scale=20, size=30)  # std dev = 20

# Step 2: Calculate variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Step 3: F-statistic (larger variance / smaller variance)
f_statistic = max(var1, var2) / min(var1, var2)

# Step 4: Degrees of freedom
df1 = len(sample1) - 1
df2 = len(sample2) - 1

# Step 5: Calculate p-value (two-tailed)
p_value = 2 * min(
    f.cdf(f_statistic, df1, df2),         
    1 - f.cdf(f_statistic, df1, df2)     
)

# Step 6: Print results
print("F-Test to Compare Variances")
print(f"Variance 1: {var1:.4f}")
print(f"Variance 2: {var2:.4f}")
print(f"F-Statistic: {f_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

# Step 7: Interpret
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis — Variances are significantly different.")
else:
    print("Result: Fail to reject the null hypothesis — Variances are not significantly different.")
```

```python
#Q17.  Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results
import numpy as np
from scipy.stats import f_oneway

# Step 1: Create sample data for 3 groups
group1 = np.random.normal(loc=50, scale=5, size=30)
group2 = np.random.normal(loc=52, scale=5, size=30)
group3 = np.random.normal(loc=60, scale=5, size=30)

# Step 2: Perform one-way ANOVA test
f_stat, p_value = f_oneway(group1, group2, group3)

# Step 3: Print the results
print("One-Way ANOVA Test:")
print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Step 4: Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis — At least one group mean is significantly different.")
else:
    print("Result: Fail to reject the null hypothesis — No significant difference between group means.")
```

```python
#Q19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA 
import numpy as np
import pandas as pd
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups):
    print("Checking ANOVA Assumptions:\n")

    # 1. Normality (Shapiro-Wilk test for each group)
    for i, group in enumerate(groups, start=1):
        stat, p = shapiro(group)
        print(f"Group {i} - Shapiro-Wilk Test:")
        print(f"  W-Statistic = {stat:.4f}, P-Value = {p:.4f}")
        if p > 0.05:
            print("  ✅ Data is likely normal.\n")
        else:
            print("  ❌ Data is not normally distributed.\n")

    # 2. Homogeneity of Variances (Levene’s Test)
    stat, p = levene(*groups)
    print("Levene’s Test for Equal Variances:")
    print(f"  F-Statistic = {stat:.4f}, P-Value = {p:.4f}")
    if p > 0.05:
        print("  ✅ Variances are likely equal.\n")
    else:
        print("  ❌ Variances are significantly different.\n")

    # 3. Independence – cannot be tested statistically; just a reminder
    print("Note: Independence is assumed based on experimental design. ✅\n")

# Example usage with 3 groups
group1 = np.random.normal(loc=50, scale=5, size=30)
group2 = np.random.normal(loc=52, scale=5, size=30)
group3 = np.random.normal(loc=54, scale=5, size=30)

check_anova_assumptions(group1, group2, group3)
```

```python
#Q20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Step 1: Create a simulated dataset
np.random.seed(42)
data = {
    'Diet': ['Low-Carb', 'Low-Fat', 'Mediterranean'] * 20,
    'Exercise': ['Yes', 'No'] * 30,
    'WeightLoss': np.random.normal(5, 2, 60)
}
df = pd.DataFrame(data)

# Step 2: Two-way ANOVA with interaction
model = ols('WeightLoss ~ C(Diet) + C(Exercise) + C(Diet):C(Exercise)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Step 3: Print ANOVA table
print("\nTwo-Way ANOVA Table (with Interaction):")
print(anova_table)

# Step 4: Visualize results using a boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(x='Diet', y='WeightLoss', hue='Exercise', data=df, palette='Set2')
plt.title('Weight Loss by Diet and Exercise')
plt.xlabel('Diet Type')
plt.ylabel('Weight Loss (kg)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Q21. Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Step 1: Set the degrees of freedom for the numerator and denominator
dfn = 5  # Degrees of freedom for the numerator
dfd = 2  # Degrees of freedom for the denominator

# Step 2: Create a range of x values (from 0 to 5, with 0.1 step size)
x = np.linspace(0, 5, 1000)

# Step 3: Calculate the F-distribution PDF (Probability Density Function)
pdf = f.pdf(x, dfn, dfd)

# Step 4: Plot the F-distribution
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label=f'F-distribution (dfn={dfn}, dfd={dfd})', color='blue')
plt.title('F-Distribution')
plt.xlabel('F-Statistic')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
#Q22. Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

# Step 1: Generate sample data for 3 groups (e.g., Group A, Group B, Group C)
np.random.seed(42)
group_A = np.random.normal(loc=50, scale=5, size=30)  # Mean=50, Std=5, n=30
group_B = np.random.normal(loc=55, scale=5, size=30)  # Mean=55, Std=5, n=30
group_C = np.random.normal(loc=60, scale=5, size=30)  # Mean=60, Std=5, n=30

# Step 2: Perform a One-Way ANOVA test
f_stat, p_value = f_oneway(group_A, group_B, group_C)

# Step 3: Print the results of the ANOVA test
print("One-Way ANOVA Test Results:")
print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis — At least one group mean is significantly different.")
else:
    print("Result: Fail to reject the null hypothesis — No significant difference between group means.")

# Step 4: Visualize the group data with boxplot
# Combine all groups into a single DataFrame
import pandas as pd
data = pd.DataFrame({
    'Score': np.concatenate([group_A, group_B, group_C]),
    'Group': ['A'] * 30 + ['B'] * 30 + ['C'] * 30
})

# Create the boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(x='Group', y='Score', data=data, palette='Set2')
plt.title('One-Way ANOVA: Comparison of Group Means')
plt.xlabel('Group')
plt.ylabel('Score')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Q23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means
import numpy as np
from scipy import stats

# Step 1: Simulate random data from a normal distribution
np.random.seed(42)  # Set seed for reproducibility
sample_size = 50
population_mean = 100
population_std = 10

# Generate data: Normal distribution with mean=100, std=10, and n=50
sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Step 2: Perform a one-sample t-test
# Hypothesis:
# H0 (Null): The sample mean = population mean (100)
# H1 (Alternative): The sample mean ≠ population mean (100)

t_stat, p_value = stats.ttest_1samp(sample_data, population_mean)

# Step 3: Print the results
print("T-Test Results:")
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Step 4: Interpretation of the results
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Result: Reject the null hypothesis — The sample mean is significantly different from the population mean.")
else:
    print("Result: Fail to reject the null hypothesis — There is no significant difference between the sample mean and the population mean.")
```

```python
#Q24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
import numpy as np
from scipy.stats import chi2

# Step 1: Simulate random data from a normal distribution
np.random.seed(42)
sample_size = 30
population_variance = 25  # Known population variance
sample_data = np.random.normal(loc=50, scale=np.sqrt(population_variance), size=sample_size)

# Step 2: Calculate the sample variance
sample_variance = np.var(sample_data, ddof=1)  # ddof=1 for sample variance (Bessel's correction)

# Step 3: Define the population variance to compare
sigma_0_squared = 25  # Hypothesized population variance (sigma_0^2)

# Step 4: Compute the Chi-square statistic
chi_square_statistic = (sample_size - 1) * sample_variance / sigma_0_squared

# Step 5: Calculate the p-value (for a two-tailed test)
alpha = 0.05  # Significance level
df = sample_size - 1  # Degrees of freedom

# P-value for the two-tailed test
p_value = 2 * min(chi2.cdf(chi_square_statistic, df), 1 - chi2.cdf(chi_square_statistic, df))

# Step 6: Print the results
print("Chi-Square Test Results:")
print(f"Chi-Square Statistic: {chi_square_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

# Step 7: Interpretation of the results
if p_value < alpha:
    print("Result: Reject the null hypothesis — The population variance is significantly different from 25.")
else:
    print("Result: Fail to reject the null hypothesis — The population variance is not significantly different from 25.")
```

```python
#Q25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups
import numpy as np
from scipy.stats import norm

# Step 1: Define the data for two groups
# Group 1
x1 = 80  # Number of successes in group 1
n1 = 100  # Total sample size for group 1
p1 = x1 / n1  # Proportion of successes in group 1

# Group 2
x2 = 70  # Number of successes in group 2
n2 = 100  # Total sample size for group 2
p2 = x2 / n2  # Proportion of successes in group 2

# Step 2: Calculate the pooled proportion (P)
P = (x1 + x2) / (n1 + n2)

# Step 3: Calculate the Z-statistic
Z = (p1 - p2) / np.sqrt(P * (1 - P) * (1/n1 + 1/n2))

# Step 4: Calculate the p-value for a two-tailed test
p_value = 2 * (1 - norm.cdf(abs(Z)))  # Two-tailed test

# Step 5: Print the results
print("Z-Test Results:")
print(f"Z-Statistic: {Z:.4f}")
print(f"P-Value: {p_value:.4f}")

# Step 6: Interpretation of the results
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Result: Reject the null hypothesis — The proportions are significantly different.")
else:
    print("Result: Fail to reject the null hypothesis — The proportions are not significantly different.")
```
